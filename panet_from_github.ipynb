```python
from owlready2 import *
# restart kernel for new owl file

panet_url = 'http://purl.org/pan-science/PaNET/PaNET.owl' # current version on github
#panet_url = '/dls/science/users/spc93/tmp/PaNET_test_PEPICO_BB_hermit.owl' # local test versions
#panet_url = '/dls/science/users/spc93/tmp/PaNET_test_PEPICO_SPC_hermit.owl'
#panet_url = '/dls/science/users/spc93/tmp/PaNET_test_IXS.owl'
#panet_url = '/dls/science/users/spc93/tmp/PaNET_31_august_2022.owl'

#technique = 'PEPICO'
#technique = 'compton scattering'
#technique = 'photoemission electron microscopy'
#technique = 'inelastic x-ray scattering'
#technique = 'x-ray photoemission electron microscopy'
#technique = 'nanofocus x-ray scattering'
technique = 'ARPES'

panet = get_ontology(panet_url).load()


for pncls in panet.classes():
    #print(pncls.label)
    if technique in pncls.label or technique in pncls.altLabel:
        panet_technique = pncls
        break

parents = panet_technique.mro()[0:-2]

print('\nParents of %s in %s:\n' % (technique, panet_url))

icount = 0        
for parent in parents:
    #print(parent)
    if parent.label == []:
        continue
    
    if not parent.altLabel == []:
        altlabels = '(Alt. Labels: ' + ', '.join(parent.altLabel) + ')'
    else:
        altlabels = ''
    
    print(parent.label[0], altlabels)
    icount += 1
        
print('\nNo. of superclasses of %s: %i' % (technique, icount))
    
```
