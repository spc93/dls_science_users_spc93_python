```python
import nexusformat

#def _find_key(tree, key, nxtype = nexusformat.nexus.tree.NXfield, attrs = [], previous_keys=[]):
#def _find_key(tree, key = '', nxtype = None, attrs = [], previous_keys=[]):
def _find_key(tree, key = '', nxtype = None, previous_keys=[]):
    
    # horrible recursive function that drills down into the nexus tree and creates a global dictionary
    #   of nexus fields/groups of a specified type and selected attributes
    #   defaults: NXfield type, no attributes added to dict
    
    global _keylist, _dictlist
    try:
        for keyval in tree.keys():
            if keyval == key or key == '':

                try:
                    #print('type: ', type(tree[keyval]) )
                    if (nxtype == None or type(tree[keyval]) == nxtype):
                    
                        dic = {'keylist': previous_keys + [keyval]}  # list of node strings
                    
                        #save all attributes to dictionary in list
                        for attrib in tree[keyval].attrs:
                            #if attrib in tree[keyval].attrs:
                            dic[attrib] = tree[keyval].attrs[attrib]
                        
                        #save value if NXfield type
                        if type(tree[keyval]) == nexusformat.nexus.tree.NXfield:
                            dic['value'] = tree[keyval]
                                
                        _dictlist += [dic]

                    
                except Exception as e:
                     print('Exception:', repr(e))
                   
                    
                _keylist += [previous_keys + [keyval]]
            #_find_key(tree[keyval], key = key, nxtype = nxtype, attrs = attrs, previous_keys = previous_keys + [keyval])
            _find_key(tree[keyval], key = key, nxtype = nxtype, previous_keys = previous_keys + [keyval])
    except:
        pass


# default to all types (nxtype = None)

   
    
    
def findkeys(keystring):
    'Return list of key sequences (lists) that end with keystring'    
    global _keylist
    _keylist=[]
    _find_key(n.nx, key = keystring)
    return _keylist

def findfieldssgrp(group, key = '', nxtype = None, attrs = []):
    # Return dictionary of nexus groups/fields of type nxtype within group (recursively)
    # optionally add specified attributes to dictionary
    
    global _keylist, _dictlist
    _keylist, _dictlist = [], []    
    #_find_key(group, key = key, nxtype = nxtype, attrs = attrs) 
    _find_key(group, key = key, nxtype = nxtype) 
    return _dictlist
```
