```python
def getClassFromLabel(label):
    for cl in list(onto.classes()):
        if label in cl.label + cl.altLabel:
            return cl
        
def getLabelsFromClassList(classlist):
    str = ''
    for cl in classlist:
        for label in cl.label + cl.altLabel:
            str += "'%s'\t" % label
        str += '\n'
    return str

def printAllSuperclassesAndAltLabels(technique):
    print(getLabelsFromClassList(getClassFromLabel(technique).mro()[:-2]))

for cl in list(onto.classes()):
   
    print('labels: ', cl.label, cl.altLabel)   
    print('Direct superclasses:')
    pprint(cl.is_a)
    print('All superclass labels:')
    print(getLabelsFromClassList(cl.mro()[:-2]))
    print
    
    
printAllSuperclassesAndAltLabels('ARPES')

    
#print(showLabelsFromClassList(getClassFromLabel('ARPES').mro()[:-2]))
```

```python
top = [getClassFromLabel('photon and neutron technique')]
maxlevel = 99
icount = 0
#could make global string and append it instead of print?

def nextClasses(level, classlist):
    global icount

    try:
        #print('%1: ', level, classlist)
        if (not classlist == []) and (level <= maxlevel):
            for cls in classlist:
                icount += 1
                print('  ' * level, cls.name)
                try:
                    (newlevel, newclasslist) = nextClasses(level+1, list(cls.subclasses()))
                except:
                    pass

            return (newlevel, newclasslist)
    except:
        pass


nextClasses(1, top)
```
