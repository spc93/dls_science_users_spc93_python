```python
import numpy as np
strain_dict={}

strain_dict['none'] = np.eye(3)

strain_dict['xx'] = np.array([
    [1, 0, 0],
    [0, 0, 0],
    [0, 0, 0]
    ])

strain_dict['yy'] = np.array([
    [0, 0, 0],
    [0, 1, 0],
    [0, 0, 0]
    ])

strain_dict['zz'] = np.array([
    [0, 0, 0],
    [0, 0, 0],
    [0, 0, 1]
    ])

strain_dict['xy'] = np.array([
    [0, 1, 0],
    [1, 0, 0],
    [0, 0, 0]
    ])

strain_dict['yz'] = np.array([
    [0, 0, 0],
    [0, 0, 1],
    [0, 1, 0]
    ])

strain_dict['zx'] = np.array([
    [0, 0, 1],
    [0, 0, 0],
    [1, 0, 0]
    ])

rnd = np.random.rand(3,3)
strain_dict['random'] = rnd + rnd.transpose() # random symmetric tensor
```

```python
import TensorScatteringClass as ten
import numpy as np
cifpath = '/home/spc93/spc_cifs/'
ciffile=cifpath+'MnF2.cif'


t=ten.TensorScatteringClass(CIFfile=ciffile, Site='Mn1');
sglist = t.sglist

for ref in [ [0,0,3], [0,3,0], [0,2,3], [0,3,2] ]:

    ref = np.array(ref)
    print('\nhkl: (%i, %i, %i)' % tuple(ref))

    for key in strain_dict:

        T = strain_dict[key]

        new_sglist = []    
        for sg in sglist:
            S = sg[0]
            if np.allclose(T, t.transform_cart(T, S)):
                new_sglist += [sg]
        t.sglist = new_sglist
        
        #assert t.isGroup(t.sglist), '=== Symmetry ops should form a group' # ok but gives warning
        t.SF_symmetry(t.sitevec, ref, t.sglist)
        print('Stress: %s\tAllowed: %r' % (key, bool(t.gen_scalar_allowed)))

    
```
