## Summary

Good results but no interference observed

Sample quality poor: peaks ~ 0.25 deg wide for high energy; ~ 0.1 deg for low energy

Orientation not along obvious direction - no obvious useful face to glue to kept initial orientation

Very nice ATS, up to ~ 10^8 cps

Multiple scattering manageable

030 showed strong quadrupole with interference with dipole

### Low energy

Non-resonant magnetic scattering up to ~ 70K cps

Inensity depending on hkl and psi but no psi scans due to odd orientation

Clear phase transition but always residual intensity aboove TN


ATS totally dominant near resonance so had to go way below to see comparable mag/ATS intensity

No clear change with magnetic field, feild cooling or flipping

Field/sample orientations very limited: Measured field dependence only at 003 with field perpendicular to 003. Might not be good orientation.

## Absence of field dependence

003 ATS ~ 10^8 cps peak, 10^7 cps E2E2
003 mag ~ 4e4 cps 6 keV
Max interference ~ 10% at E2

Field cooling #923220 etc: No clear difference with field sign at various energies around resonace.
No clear difference below resonance where amplitudes are more equal.
003 psi=38: polarization for E1E1 resonance almost all sigma-sigma - this will not have helped.


Below resonance - interference elimited due to amplitudes being almost out of phase
Likely that there is no change in domains with field perp to 003.
